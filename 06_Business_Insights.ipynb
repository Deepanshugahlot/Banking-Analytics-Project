# 📈 Banking Analytics Project – Business Insights

## Notebook 06: Business Insights & Strategic Recommendations

This notebook summarizes the key findings from the EDA and presents actionable business recommendations.

## Executive Summary

- **150,000 Customers**
- **240,000 Accounts**
- **2M+ Transactions**
- Loan Portfolio
- Customer Complaints
- Branch Performance
- Merchant Activity

## 👥 Customer Insights

- 📍 **Regular** is the largest customer segment (**61,393 customers**) but maintains the **lowest average balance (₹30,278)**.
- 💎 **Premium** customers (**33,196**) maintain the **highest average balance (₹2,48,885)**.
- ⚠️ **17.6% of customers (26,430)** belong to the High Risk category.
- 🏦 Savings Accounts dominate the portfolio, creating cross-selling opportunities.

## 💳 Transaction Insights

- 📈 Transaction volume increased steadily from 2022–2025.
- 💰 Deposits (14,17,776) significantly exceed Purchases (6,60,056).
- 🔄 Bank Transfer is the most frequently used transaction channel.
- 🛒 Groceries, Online Shopping, and Dining are the top merchant categories.
- 🚨 Fraud remains a small percentage of transactions but shows a gradual increase.

## 🏠 Loan Insights

- Personal Loans are issued most frequently.
- Home Loans have the highest average loan amount.
- Higher-risk customers show greater default proportions.
- EMI increases proportionally with loan amount.

## 📞 Complaint Insights

- Approximately 70% of complaints are resolved.
- Phone is the primary complaint channel.
- Fraud and Transaction Dispute complaints require the longest resolution times.
- Resolution times are relatively consistent across standard complaint types.

## 🏢 Branch & Merchant Insights

- Kochi Branch 2 processes the highest transaction volume.
- Branch performance varies across locations.
- Salary Credit generates high transaction value.
- Kochi, Hyderabad, Jaipur, and Kolkata are major activity hubs.

## 💼 Strategic Recommendations

1. Retain Premium and Loyal customers.
2. Cross-sell Current and Credit Accounts.
3. Strengthen fraud monitoring.
4. Improve complaint resolution.
5. Benchmark top-performing branches.
6. Increase digital banking adoption.

## ✅ Final Conclusion

This Banking Analytics Project transformed raw banking data into business intelligence using Python, Pandas, Matplotlib, and Tableau.

The analysis identified customer behavior, transaction trends, loan risk, complaint patterns, and branch performance, providing actionable insights to support strategic decision-making.
