## 12. Foreign Key Validation

```python
relationships = [
    # Accounts
    ("Accounts → Customers", accounts, "customer_id", customers, "customer_id"),
    ("Accounts → Branches", accounts, "branch_id", branches, "branch_id"),

    # Cards
    ("Cards → Accounts", cards, "account_id", accounts, "account_id"),

    # Transactions
    ("Transactions → Accounts", transactions, "account_id", accounts, "account_id"),

    # Ignore NULL card_id values (Salary deposits, etc.)
    ("Transactions → Cards",
     transactions.dropna(subset=["card_id"]),
     "card_id",
     cards,
     "card_id"),

    # Ignore NULL merchant_id values (Internal transfers, salary credits, etc.)
    ("Transactions → Merchants",
     transactions.dropna(subset=["merchant_id"]),
     "merchant_id",
     merchants,
     "merchant_id"),

    ("Transactions → Date",
     transactions,
     "date_id",
     dates,
     "date_id"),

    # Loans
    ("Loans → Customers",
     loans,
     "customer_id",
     customers,
     "customer_id"),

    # Complaints
    ("Complaints → Customers",
     complaints,
     "customer_id",
     customers,
     "customer_id"),

    ("Complaints → Accounts",
     complaints,
     "account_id",
     accounts,
     "account_id"),

    # Ignore complaints that are not linked to a transaction
    ("Complaints → Transactions",
     complaints.dropna(subset=["linked_transaction_id"]),
     "linked_transaction_id",
     transactions,
     "transaction_id"),

    ("Complaints → Date",
     complaints,
     "date_id",
     dates,
     "date_id")
]



fk_results = []

for relation, child_df, child_col, parent_df, parent_col in relationships:

    invalid_records = child_df.loc[
        ~child_df[child_col].isin(parent_df[parent_col])
    ].shape[0]

    fk_results.append({
        "Relationship": relation,
        "Rows Checked": len(child_df),
        "Invalid Records": invalid_records,
        "Status": "✅ Pass" if invalid_records == 0 else "❌ Fail"
    })

fk_results = pd.DataFrame(fk_results)

fk_results
```

## 14. Observations

- The project contains 9 related tables following a star schema (Customers, Accounts, Cards, Branches, Merchants, Date as dimensions; Transactions, Loans, Complaints as facts).
- All 12 foreign key relationships checked (Accounts, Cards, Transactions, Loans, Complaints against their parent tables) passed with 0 invalid records.
- No duplicate primary keys exist in any dimension table. Transactions has 24,978 duplicate `transaction_id` values, of which 23,372 are full-row duplicates and the remaining ~1,600 differ in at least one other column. This pattern is typical of duplicate ETL re-ingestion and will be handled in the Cleaning phase.
- Missing values fall into two groups: (1) genuinely optional fields such as `gender`, `occupation`, `education`, and `income_bracket` in Customers (1.5%-3% missing each), and (2) structurally explainable nulls such as `card_id` (51.96% missing — cash/UPI/internal transfers have no card) and `merchant_id` (1.5% missing — salary credits and internal transfers have no merchant). `linked_transaction_id` is missing for 80.91% of complaints, consistent with complaints that aren't tied to a specific transaction (e.g. service or KYC issues), and `resolution_days` is missing for 29.90% of complaints, consistent with complaints still open/unresolved.
- `transactions.amount` ranges from roughly -1.97 crore to +2.68 crore, and customer age/income also show wide ranges (age up to 85, monthly income up to ₹3,99,999). These extremes need to be investigated in the Cleaning phase to confirm whether they are genuine outliers, refunds/reversals, or data quality issues.
- Date columns (`signup_date`, `open_date`, `issue_date`, `disbursement_date`, `full_date`) were converted from string to datetime for future time-based analysis.
- The dataset structure is validated and ready for the Data Cleaning phase, with duplicate transactions and the amount/age/income outliers above as the main items to address first.
